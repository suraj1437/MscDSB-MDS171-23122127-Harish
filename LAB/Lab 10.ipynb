```python
class Student:
    def __init__(self, student_id, name, age):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.marks = []

    def add_marks(self, subject, mark):
        self.marks.append((subject, mark))

    def calculate_average_mark(self):
        total_marks = sum(mark for _, mark in self.marks)
        return total_marks / len(self.marks) if self.marks else 0

    def display_student_info(self):
        print("Student ID: " + str(self.student_id))
        print("Name: " + self.name)
        print("Age: " + str(self.age))
        if self.marks:
            print("Marks:")
            for subject, mark in self.marks:
                print(subject + ": " + str(mark))
            print("Average Mark: " + str(self.calculate_average_mark()))
        else:
            print("No marks available for this student.")

# Menu-driven program
students = []

while True:
    print("\nMenu:")
    print("1. Add Student")
    print("2. Enter Marks")
    print("3. Display Student Information")
    print("4. Exit")
    
    choice = input("Enter your choice (1/2/3/4): ")
    
    if choice == "1":
        student_id = input("Enter Student ID: ")
        name = input("Enter Student Name: ")
        age = int(input("Enter Student Age: "))
        student = Student(student_id, name, age)
        students.append(student)
        print("Student added successfully.")
    
    elif choice == "2":
        if not students:
            print("No students added. Please add a student first.")
            continue
        student_id = input("Enter Student ID for marks entry: ")
        found = False
        for student in students:
            if student.student_id == student_id:
                subject = input("Enter Subject: ")
                mark = float(input("Enter Mark: "))
                student.add_marks(subject, mark)
                found = True
                print("Marks entered successfully.")
                break
        if not found:
            print("Student with ID " + student_id + " not found.")
    
    elif choice == "3":
        if not students:
            print("No students added. Please add a student first.")
            continue
        student_id = input("Enter Student ID for information: ")
        found = False
        for student in students:
            if student.student_id == student_id:
                student.display_student_info()
                found = True
                break
        if not found:
            print("Student with ID " + student_id + " not found.")
    
    elif choice == "4":
        print("Exiting the program. Goodbye!")
        break
    
    else:
        print("Invalid choice. Please select a valid option (1/2/3/4).")
```
