```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("StudentData.csv")

# Filter the data for only those who have asked for money due to poor money management skills
df = df[df['Have you asked your parents for money due to poor money management skills? '] == 'Yes']

# Create a new dataframe with the count of students for each branch
count_df = df.groupby('Select Your Branch.').size().reset_index(name='count')

# Calculate percentage
count_df['percentage'] = count_df['count'] / count_df['count'].sum() * 100

# Create a pie chart
plt.pie(count_df['count'], labels=count_df['Select Your Branch.'], autopct='%1.1f%%')
plt.title('Percentage of students who have asked for money by branch')

plt.show()
```
