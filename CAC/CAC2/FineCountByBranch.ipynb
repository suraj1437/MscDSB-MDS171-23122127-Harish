```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from CSV file
data = pd.read_csv('StudentData.csv')

# Group by branch and Fine Paid and count number of responses
Fine_Count = data.groupby(['Select Your Branch.', 'Have you paid fine in Christ? ']).size().reset_index(name='count')
print(Fine_Count)
# Pivot table and create stacked bar chart
Fine_Count = Fine_Count.pivot(index='Select Your Branch.', columns='Have you paid fine in Christ? ', values='count')

Fine_Count.plot(kind='bar', stacked=True)
plt.title('Fine Count')
plt.xlabel('Branch')
plt.ylabel('Number of Responses')
plt.legend(title='Fine Amount', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
