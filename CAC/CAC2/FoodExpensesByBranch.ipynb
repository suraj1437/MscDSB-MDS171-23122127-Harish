```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 10, 'figure.figsize': (17, 8)})

# Load data from CSV file
data = pd.read_csv('StudentData.csv')

# Create box plot
data.boxplot(column='How much do you spend on your Food? (includes Breakfast, Lunch, Dinner, Tea/Coffee', by='Select Your Branch.')
plt.suptitle('')
plt.title('Food Expenses by Branch')
plt.xlabel('Branch')
plt.ylabel('Food Expenses (Rs.)')
plt.show()
```
